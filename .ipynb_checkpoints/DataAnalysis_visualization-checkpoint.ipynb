```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')

# í•œê¸€ í°íŠ¸ ì„¤ì •
plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['axes.unicode_minus'] = False

print("="*60)
print("í•œêµ­ì „ë ¥ê³µì‚¬ ì „ë ¥ ì‚¬ìš©íŒ¨í„´ ë³€ë™ê³„ìˆ˜ ê°œë°œ í”„ë¡œì íŠ¸")
print("Step 1: ë°ì´í„° íƒìƒ‰ ë° ê¸°ë³¸ ë¶„ì„")
print("="*60)

# ============================================================================
# 1. ëŒ€ìƒê³ ê° ë°ì´í„° ë¡œë”© ë° íƒìƒ‰
# ============================================================================

print("\n[1ë‹¨ê³„] ëŒ€ìƒê³ ê° ë°ì´í„° ë¶„ì„")
print("-" * 40)

# ê³ ê° ë°ì´í„° ë¡œë”© (ì‹¤ì œ í™˜ê²½ì—ì„œëŠ” íŒŒì¼ ê²½ë¡œ ìˆ˜ì • í•„ìš”)
try:
    # customer_df = pd.read_excel('ì œ13íšŒ ì‚°ì—…ë¶€ ê³µëª¨ì „ ëŒ€ìƒê³ ê°.xlsx')
    # ì„ì‹œë¡œ ìƒ˜í”Œ ë°ì´í„° ìƒì„± (ì‹¤ì œ ë¶„ì„ì‹œì—ëŠ” ìœ„ ì½”ë“œ ì‚¬ìš©)
    customer_df = pd.DataFrame({
        'ìˆœë²ˆ': range(1, 3001),
        'ê³ ê°ë²ˆí˜¸': [f'CUST_{i:04d}' for i in range(1, 3001)],
        'ê³„ì•½ì „ë ¥': np.random.choice([100, 200, 300, 500, 700, 900], 3000),
        'ê³„ì•½ì¢…ë³„': np.random.choice([222, 226, 311, 322, 726], 3000),
        'ì‚¬ìš©ìš©ë„': np.random.choice(['02', '09'], 3000),
        'ì£¼ìƒì‚°í’ˆ': ['ì œí’ˆA', 'ì œí’ˆB', 'ì œí’ˆC', '', 'ì œí’ˆD'] * 600,
        'ì‚°ì—…ë¶„ë¥˜(ì†Œ)': ['ì—…ì¢…1', 'ì—…ì¢…2', '#N/A', 'ì—…ì¢…3', 'ì—…ì¢…4'] * 600
    })
    
    print(f"âœ“ ê³ ê° ë°ì´í„° ë¡œë”© ì™„ë£Œ: {len(customer_df):,}ê±´")
    
except Exception as e:
    print(f"âœ— ê³ ê° ë°ì´í„° ë¡œë”© ì‹¤íŒ¨: {e}")

# ê¸°ë³¸ ì •ë³´ í™•ì¸
print(f"\nğŸ“‹ ê³ ê° ë°ì´í„° ê¸°ë³¸ ì •ë³´:")
print(f"   - ì´ ê³ ê° ìˆ˜: {len(customer_df):,}ëª…")
print(f"   - ì»¬ëŸ¼ ìˆ˜: {len(customer_df.columns)}ê°œ")
print(f"   - ë©”ëª¨ë¦¬ ì‚¬ìš©ëŸ‰: {customer_df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")

# ì»¬ëŸ¼ë³„ ë°ì´í„° íƒ€ì… ë° ê²°ì¸¡ì¹˜ í™•ì¸
print(f"\nğŸ“Š ì»¬ëŸ¼ë³„ ìƒì„¸ ì •ë³´:")
for col in customer_df.columns:
    null_count = customer_df[col].isnull().sum()
    null_pct = (null_count / len(customer_df)) * 100
    unique_count = customer_df[col].nunique()
    print(f"   {col:15s}: {customer_df[col].dtype:10s} | ê²°ì¸¡ì¹˜: {null_count:4d}ê°œ ({null_pct:5.1f}%) | ê³ ìœ ê°’: {unique_count:4d}ê°œ")

# ê³„ì•½ì¢…ë³„ ë¶„í¬
print(f"\nğŸ“ˆ ê³„ì•½ì¢…ë³„ ë¶„í¬:")
contract_counts = customer_df['ê³„ì•½ì¢…ë³„'].value_counts().sort_index()
for contract_type, count in contract_counts.items():
    pct = (count / len(customer_df)) * 100
    # ê³„ì•½ì¢…ë³„ ì„¤ëª… ë§¤í•‘
    contract_desc = {
        222: "ì¼ë°˜ìš©(ê°‘)||ê³ ì••A",
        226: "ì¼ë°˜ìš©(ì„) ê³ ì••A", 
        311: "ì‚°ì—…ìš©(ê°‘) ì €ì••",
        322: "ì‚°ì—…ìš©(ê°‘)||ê³ ì••A",
        726: "ì‚°ì—…ìš©(ì„) ê³ ì••A"
    }
    desc = contract_desc.get(contract_type, "ë¯¸ë¶„ë¥˜")
    print(f"   {contract_type}: {count:4d}ëª… ({pct:5.1f}%) - {desc}")

# ì‚¬ìš©ìš©ë„ ë¶„í¬
print(f"\nğŸ“ˆ ì‚¬ìš©ìš©ë„ ë¶„í¬:")
usage_counts = customer_df['ì‚¬ìš©ìš©ë„'].value_counts().sort_index()
for usage_type, count in usage_counts.items():
    pct = (count / len(customer_df)) * 100
    usage_desc = {"02": "ìƒì—…ìš©", "09": "ê´‘ê³µì—…ìš©"}
    desc = usage_desc.get(usage_type, "ë¯¸ë¶„ë¥˜")
    print(f"   {usage_type}: {count:4d}ëª… ({pct:5.1f}%) - {desc}")

# ê³„ì•½ì „ë ¥ ë¶„í¬
print(f"\nğŸ“ˆ ê³„ì•½ì „ë ¥ ë¶„í¬:")
power_stats = customer_df['ê³„ì•½ì „ë ¥'].describe()
print(f"   í‰ê· : {power_stats['mean']:6.1f} | ì¤‘ìœ„ìˆ˜: {power_stats['50%']:6.1f}")
print(f"   ìµœì†Œ: {power_stats['min']:6.1f} | ìµœëŒ€: {power_stats['max']:6.1f}")
print(f"   í‘œì¤€í¸ì°¨: {power_stats['std']:6.1f}")

# ============================================================================
# 2. LP ë°ì´í„° ìƒ˜í”Œ ìƒì„± ë° íƒìƒ‰ (ì‹¤ì œ í™˜ê²½ì—ì„œëŠ” íŒŒì¼ ë¡œë”©)
# ============================================================================

print(f"\n\n[2ë‹¨ê³„] LP ë°ì´í„° ë¶„ì„")
print("-" * 40)

# LP ë°ì´í„° ìƒ˜í”Œ ìƒì„± (ì‹¤ì œ ë¶„ì„ì‹œì—ëŠ” CSV íŒŒì¼ë“¤ì„ ë¡œë”©)
print("ğŸ“ LP ë°ì´í„° ìƒ˜í”Œ ìƒì„± ì¤‘...")

# ì‹œê°„ ë²”ìœ„ ì„¤ì •
start_date = datetime(2022, 3, 16, 0, 15)
end_date = datetime(2022, 4, 1, 0, 0)
time_range = pd.date_range(start=start_date, end=end_date, freq='15min')

# ê³ ê°ë³„ LP ë°ì´í„° ìƒ˜í”Œ ìƒì„± (ë©”ëª¨ë¦¬ íš¨ìœ¨ì„±ì„ ìœ„í•´ ì¼ë¶€ ê³ ê°ë§Œ)
sample_customers = customer_df['ê³ ê°ë²ˆí˜¸'].head(100).tolist()  # ìƒ˜í”Œ 100ëª…
lp_data_list = []

for customer_id in sample_customers[:10]:  # ë” ì‘ì€ ìƒ˜í”Œë¡œ ì‹œì‘
    for timestamp in time_range[:96]:  # 1ì¼ì¹˜ë§Œ ìƒ˜í”Œ
        # ì‹¤ì œì ì¸ ì „ë ¥ ì‚¬ìš© íŒ¨í„´ ì‹œë®¬ë ˆì´ì…˜
        hour = timestamp.hour
        
        # ì‹œê°„ëŒ€ë³„ ê¸°ë³¸ ë¶€í•˜ íŒ¨í„´ (ì—…ë¬´ì‹œê°„ ë†’ìŒ)
        if 9 <= hour <= 18:
            base_load = np.random.normal(50, 10)
        elif 19 <= hour <= 22:
            base_load = np.random.normal(30, 8)
        else:
            base_load = np.random.normal(15, 5)
        
        # ìŒìˆ˜ ë°©ì§€
        base_load = max(0, base_load)
        
        lp_data_list.append({
            'ëŒ€ì²´ê³ ê°ë²ˆí˜¸': customer_id,
            'LP ìˆ˜ì‹ ì¼ì': timestamp.strftime('%Y-%m-%d %H:%M'),
            'ìˆœë°©í–¥ ìœ íš¨ì „ë ¥': round(base_load + np.random.normal(0, 2), 1),
            'ì§€ìƒë¬´íš¨': round(base_load * 0.3 + np.random.normal(0, 1), 1),
            'ì§„ìƒë¬´íš¨': round(base_load * 0.2 + np.random.normal(0, 0.5), 1),
            'í”¼ìƒì „ë ¥': round(base_load * 1.1 + np.random.normal(0, 1), 1)
        })

lp_df = pd.DataFrame(lp_data_list)

# ìŒìˆ˜ê°’ ì²˜ë¦¬
numeric_cols = ['ìˆœë°©í–¥ ìœ íš¨ì „ë ¥', 'ì§€ìƒë¬´íš¨', 'ì§„ìƒë¬´íš¨', 'í”¼ìƒì „ë ¥']
for col in numeric_cols:
    lp_df[col] = lp_df[col].clip(lower=0)

print(f"âœ“ LP ë°ì´í„° ìƒ˜í”Œ ìƒì„± ì™„ë£Œ: {len(lp_df):,}ê±´")

# LP ë°ì´í„° ê¸°ë³¸ ì •ë³´
print(f"\nğŸ“‹ LP ë°ì´í„° ê¸°ë³¸ ì •ë³´:")
print(f"   - ì´ ë ˆì½”ë“œ ìˆ˜: {len(lp_df):,}ê±´")
print(f"   - ëŒ€ìƒ ê³ ê° ìˆ˜: {lp_df['ëŒ€ì²´ê³ ê°ë²ˆí˜¸'].nunique():,}ëª…")
print(f"   - ì‹œê°„ ë²”ìœ„: {lp_df['LP ìˆ˜ì‹ ì¼ì'].min()} ~ {lp_df['LP ìˆ˜ì‹ ì¼ì'].max()}")
print(f"   - ë©”ëª¨ë¦¬ ì‚¬ìš©ëŸ‰: {lp_df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")

# ìˆ˜ì¹˜ ë°ì´í„° ê¸°ë³¸ í†µê³„
print(f"\nğŸ“Š ì „ë ¥ ë°ì´í„° ê¸°ë³¸ í†µê³„:")
print(lp_df[numeric_cols].describe().round(2))

# ============================================================================
# 3. ë°ì´í„° í’ˆì§ˆ ê²€ì¦
# ============================================================================

print(f"\n\n[3ë‹¨ê³„] ë°ì´í„° í’ˆì§ˆ ê²€ì¦")
print("-" * 40)

# ê²°ì¸¡ì¹˜ í™•ì¸
print("ğŸ” ê²°ì¸¡ì¹˜ í˜„í™©:")
for df_name, df in [("ê³ ê° ë°ì´í„°", customer_df), ("LP ë°ì´í„°", lp_df)]:
    null_counts = df.isnull().sum()
    if null_counts.sum() > 0:
        print(f"   {df_name}:")
        for col, null_count in null_counts[null_counts > 0].items():
            print(f"     - {col}: {null_count:,}ê±´")
    else:
        print(f"   {df_name}: ê²°ì¸¡ì¹˜ ì—†ìŒ âœ“")

# ì¤‘ë³µ ë°ì´í„° í™•ì¸
print(f"\nğŸ” ì¤‘ë³µ ë°ì´í„° í˜„í™©:")
customer_dups = customer_df.duplicated().sum()
lp_dups = lp_df.duplicated(['ëŒ€ì²´ê³ ê°ë²ˆí˜¸', 'LP ìˆ˜ì‹ ì¼ì']).sum()
print(f"   ê³ ê° ë°ì´í„° ì¤‘ë³µ: {customer_dups:,}ê±´")
print(f"   LP ë°ì´í„° ì¤‘ë³µ: {lp_dups:,}ê±´")

# ë°ì´í„° ì—°ê³„ì„± í™•ì¸
print(f"\nğŸ” ë°ì´í„° ì—°ê³„ì„± í™•ì¸:")
customer_ids = set(customer_df['ê³ ê°ë²ˆí˜¸'].unique())
lp_customer_ids = set(lp_df['ëŒ€ì²´ê³ ê°ë²ˆí˜¸'].unique())

matched_customers = customer_ids.intersection(lp_customer_ids)
unmatched_in_customer = customer_ids - lp_customer_ids
unmatched_in_lp = lp_customer_ids - customer_ids

print(f"   ì—°ê³„ ê°€ëŠ¥ ê³ ê°: {len(matched_customers):,}ëª…")
print(f"   ê³ ê° ë°ì´í„°ë§Œ ì¡´ì¬: {len(unmatched_in_customer):,}ëª…")
print(f"   LP ë°ì´í„°ë§Œ ì¡´ì¬: {len(unmatched_in_lp):,}ëª…")

# ============================================================================
# 4. ê¸°ë³¸ ì‹œê°í™”
# ============================================================================

print(f"\n\n[4ë‹¨ê³„] ê¸°ë³¸ ì‹œê°í™”")
print("-" * 40)

# ì‹œê°í™” ìƒì„±
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
fig.suptitle('í•œêµ­ì „ë ¥ê³µì‚¬ ê³ ê° ë° LP ë°ì´í„° ê¸°ë³¸ ë¶„ì„', fontsize=16, y=0.98)

# 1. ê³„ì•½ì¢…ë³„ ë¶„í¬
ax1 = axes[0, 0]
contract_counts.plot(kind='bar', ax=ax1, color='skyblue')
ax1.set_title('ê³„ì•½ì¢…ë³„ ê³ ê° ë¶„í¬')
ax1.set_xlabel('ê³„ì•½ì¢…ë³„')
ax1.set_ylabel('ê³ ê° ìˆ˜')
ax1.tick_params(axis='x', rotation=45)

# 2. ì‚¬ìš©ìš©ë„ ë¶„í¬
ax2 = axes[0, 1]
usage_counts.plot(kind='pie', ax=ax2, autopct='%1.1f%%', colors=['lightcoral', 'lightgreen'])
ax2.set_title('ì‚¬ìš©ìš©ë„ ë¶„í¬')
ax2.set_ylabel('')

# 3. ê³„ì•½ì „ë ¥ ë¶„í¬
ax3 = axes[0, 2]
customer_df['ê³„ì•½ì „ë ¥'].hist(bins=20, ax=ax3, color='gold', alpha=0.7)
ax3.set_title('ê³„ì•½ì „ë ¥ ë¶„í¬')
ax3.set_xlabel('ê³„ì•½ì „ë ¥')
ax3.set_ylabel('ë¹ˆë„')

# 4. ì‹œê°„ëŒ€ë³„ í‰ê·  ì „ë ¥ ì‚¬ìš©ëŸ‰
if len(lp_df) > 0:
    lp_df['LP ìˆ˜ì‹ ì¼ì'] = pd.to_datetime(lp_df['LP ìˆ˜ì‹ ì¼ì'])
    lp_df['ì‹œê°„'] = lp_df['LP ìˆ˜ì‹ ì¼ì'].dt.hour
    hourly_avg = lp_df.groupby('ì‹œê°„')['ìˆœë°©í–¥ ìœ íš¨ì „ë ¥'].mean()
    
    ax4 = axes[1, 0]
    hourly_avg.plot(kind='line', ax=ax4, marker='o', color='green')
    ax4.set_title('ì‹œê°„ëŒ€ë³„ í‰ê·  ì „ë ¥ ì‚¬ìš©ëŸ‰')
    ax4.set_xlabel('ì‹œê°„')
    ax4.set_ylabel('í‰ê·  ìœ íš¨ì „ë ¥')
    ax4.grid(True, alpha=0.3)

# 5. ì „ë ¥ ë°ì´í„° ë°•ìŠ¤í”Œë¡¯
ax5 = axes[1, 1]
lp_df[['ìˆœë°©í–¥ ìœ íš¨ì „ë ¥', 'ì§€ìƒë¬´íš¨', 'ì§„ìƒë¬´íš¨']].boxplot(ax=ax5)
ax5.set_title('ì „ë ¥ ë°ì´í„° ë¶„í¬')
ax5.set_ylabel('ì „ë ¥ê°’')
ax5.tick_params(axis='x', rotation=45)

# 6. ê³ ê°ë³„ í‰ê·  ì „ë ¥ ì‚¬ìš©ëŸ‰ (ìƒìœ„ 10ëª…)
ax6 = axes[1, 2]
customer_avg = lp_df.groupby('ëŒ€ì²´ê³ ê°ë²ˆí˜¸')['ìˆœë°©í–¥ ìœ íš¨ì „ë ¥'].mean().sort_values(ascending=False).head(10)
customer_avg.plot(kind='bar', ax=ax6, color='purple', alpha=0.7)
ax6.set_title('ê³ ê°ë³„ í‰ê·  ì „ë ¥ ì‚¬ìš©ëŸ‰ (ìƒìœ„ 10ëª…)')
ax6.set_xlabel('ê³ ê°ë²ˆí˜¸')
ax6.set_ylabel('í‰ê·  ìœ íš¨ì „ë ¥')
ax6.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

print("âœ“ ê¸°ë³¸ ì‹œê°í™” ì™„ë£Œ")

# ============================================================================
# 5. ë‹¤ìŒ ë‹¨ê³„ ì¤€ë¹„
# ============================================================================

print(f"\n\n[ì™„ë£Œ] Step 1 ë°ì´í„° íƒìƒ‰ ì™„ë£Œ")
print("="*60)
print("âœ… ë‹¤ìŒ ë‹¨ê³„:")
print("   1. ë°ì´í„° ì „ì²˜ë¦¬ ë° ì •ì œ")
print("   2. ë³€ë™ê³„ìˆ˜ ì •ì˜ ë° ê³„ì‚°")
print("   3. ê³ ê°ë³„ íŒ¨í„´ ë¶„ì„")
print("   4. ì˜ˆì¸¡ ëª¨ë¸ ê°œë°œ")
print("="*60)
```
